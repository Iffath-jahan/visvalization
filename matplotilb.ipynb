```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.random.rand(50)
Y=np.random.rand(50)
```

```python
x
```

```python
Y
```

```python
plt.plot(x,Y)
```

```python
plt.scatter(x,Y)
```

```python
plt.scatter(x,Y,c="r")
plt.xlabel("this is x values")
plt.ylabel("this is y values")
plt.title("this my x and y graph")
```

```python
plt.scatter(x,Y,c="#6d0d2d",alpha=.9)
plt.xlabel("this is x values")
plt.ylabel("this is y values")
plt.title("this my x and y graph")
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,Y,c="#6d0d2d",alpha=.9,marker='x')
plt.xlabel("this is x values")
plt.ylabel("this is y values")
plt.title("this my x and y graph")
plt.grid()
```

```python
x=["a","b","c","d","e"]
y=np.random.rand(5)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(5,3))
plt.bar(x,y,color="#1e0d6d")
plt.xlabel("representing my numerical cat data")
plt.xlabel("representing my numerical num data")
plt.title("this is my bar graph")
plt.grid()
```

```python
plt.figure(figsize=(5,3))
plt.barh(x,y,color="#1e0d6d")
plt.xlabel("representing my numerical cat data")
plt.xlabel("representing my numerical num data")
plt.title("this is my horizontal  bar graph")
plt.grid()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
fig=plt.figure()
ax=fig.add_subplot(projection ='3d')
plt.scatter(x,y,z,alpha=.9,c='#191514')
plt.xlabel("this my x axs")
plt.ylabel("this is y axis")
plt.show()
```
